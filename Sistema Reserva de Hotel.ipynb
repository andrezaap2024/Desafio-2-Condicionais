***Sistema de Reservas de Hotel:***

***Você foi contratado(a) para desenvolver uma parte do sistema de um hotel. O objetivo é criar um sistema que gerencie reservas de quartos e o pagamento das diárias***.

- *Cadastro de Cliente:*

*O sistema deve permitir que o usuário "cadastre" o nome e a idade de até 3 clientes.*

***O sistema deve oferecer 3 tipos de quartos:*** 

***"Simples", "Duplo" e "Luxo".***

***Cada cliente deve escolher um quarto para sua estadia.
O preço da diária varia conforme o tipo de quarto:
Simples: R$ 100,00 por dia.
Duplo: R$ 150,00 por dia.
Luxo: R$ 250,00 por dia.***

- ***Cálculo da Estadia:***

***O usuário deve informar quantos dias cada cliente ficará no hotel.
O sistema deve calcular o valor total da estadia para cada cliente, considerando o tipo de quarto e a quantidade de dias.***

Exemplo: 

 ***valor_cliente3 = preco_duplo * cliente3_dias***

*Pagamento:*

*O sistema deve exibir o valor total a ser pago por cada cliente após a aplicação do desconto.*

*Regras Adicionais:
Utilize apenas variáveis, condicionais (if, elif, else) e listas para resolver o desafio.*

***O sistema não deve usar loops (for, while) nem funções.**
O código deve considerar todos os casos possíveis de escolha dos clientes.*

```python
cadastro_clientes = []
tipos_quartos = ["Simples", "Duplo", "Luxo"]
preco_quartos = [100, 150, 250]
carrinho = []
meus_valores = []
forma_pagamento = ["Cartão de Crédito", "Pix", "Boleto"]
tipo_pagamento = []
qtdedias_reserva = []

nome_funcionario = str(input("Digite o nome do funcionário"))
print(f"{nome_funcionario} Seja Bem Vindo ao Sistema de Reserva do Hotel Time Square - NY SA !!!!")
 
    
nome_completo1 = input("Digite o nome completo cliente1")
idade_cliente1  = input("Digite a idade do cliente1:")
cadastro_clientes.append({ "nome_completo": nome_completo1, "idade_cliente": idade_cliente1})
    
nome_completo2 = input("Digite o nome completo cliente2")
idade_cliente2  = input("Digite a idade do cliente2:")
cadastro_clientes.append({ "nome_completo": nome_completo2, "idade_cliente": idade_cliente2})
    
nome_completo3 = input("Digite o nome completo cliente3")
idade_cliente3  = input("Digite a idade do cliente3:")
cadastro_clientes.append({ "nome_completo": nome_completo3, "idade_cliente": idade_cliente3})
    
   
  

print("\nDados pessoais cadastrados:") 
print(f"Nome Completo: {cadastro_clientes[0]['nome_completo']}, Idade: {cadastro_clientes[0]['idade_cliente']}")
print(f"Nome Completo: {cadastro_clientes[1]['nome_completo']}, Idade: {cadastro_clientes[1]['idade_cliente']}")
print(f"Nome Completo: {cadastro_clientes[2]['nome_completo']}, Idade: {cadastro_clientes[2]['idade_cliente']}")

print (f"{nome_funcionario}, Agora que você já realizou o cadastro dos clientes você irá selecionar o tipo de quarto em que o cliente deseja se hospedar")

cliente1 = int(input(f"""Digite  o código do tipo de quarto que o cliente1 deseja se hospedar
0 - Simples
1 - Duplo 
2 - Luxo                
"""))

cliente2 = int(input(f"""Digite  o código do tipo de quarto que o cliente2 deseja se hospedar
0 - Simples
1 - Duplo 
2 - Luxo                
"""))

cliente3 = int(input(f"""Digite  o código do tipo de quarto que o cliente3 deseja se hospedar
0 - Simples
1 - Duplo 
2 - Luxo                
"""))

# Qtde Dias Reserva
qtde_dias_cliente1 = int(input(f"{nome_funcionario} informe a quantidade de dias que o cliente1 irá ficar no hotel"))
qtdedias_reserva.append(qtde_dias_cliente1)
                               
qtde_dias_cliente2 = int(input(f"{nome_funcionario} informe a quantidade de dias que o cliente2 irá ficar no hotel"))
qtdedias_reserva.append(qtde_dias_cliente2)
                               
qtde_dias_cliente3 = int(input(f"{nome_funcionario} informe a quantidade de dias que o cliente3 irá ficar no hotel"))
qtdedias_reserva.append(qtde_dias_cliente3)
                               
carrinho = [tipos_quartos[0], tipos_quartos[1], tipos_quartos[2]]
meus_valores = [preco_quartos[0], preco_quartos[1], preco_quartos[2]]

vltotal_reservacliente1 = preco_quartos[cliente1] * qtdedias_reserva[0]
vltotal_reservacliente2 = preco_quartos[cliente2] * qtdedias_reserva[1]
vltotal_reservacliente3 = preco_quartos[cliente3] * qtdedias_reserva[2]

# Soma total de todas as reservas 
soma = vltotal_reservacliente1 + vltotal_reservacliente2 + vltotal_reservacliente3 


# desconto individual para cada cliente hospedado

desconto_cliente1 = vltotal_reservacliente1 * 0.90 
desconto_cliente2 = vltotal_reservacliente2 * 0.90 
desconto_cliente3 = vltotal_reservacliente3 * 0.90

pagamento = int(input(f"""  seus produtos já estão no cesto de pamento agora selecione a forma de pagamento
0 - Cartão de Crédito
1 - Pix
2 - Boleto
"""))

print(f"""
============================================================ Nota Fiscal ================================================================================
    {cadastro_clientes[0]['nome_completo']} - reservou o quarto {carrinho[0]} - onde o valor da diária é de R$ {meus_valores[0]:.2f} - por {qtdedias_reserva[0]} dias
    {cadastro_clientes[1]['nome_completo']} - reservou o quarto {carrinho[1]} - onde o valor da diária é de R$ {meus_valores[1]:.2f} - por {qtdedias_reserva[1]} dias
    {cadastro_clientes[2]['nome_completo']} - reservou o quarto {carrinho[2]} - onde o valor da diária é de R$ {meus_valores[2]:.2f} - por {qtdedias_reserva[2]} dias

======================================================= Valor Total da Compra ===========================================================================
    {cadastro_clientes[0]['nome_completo']} irá pagar o valor Total: R$ {vltotal_reservacliente1:.2f} no quarto {carrinho[0]} por {qtdedias_reserva[0]} dias
    {cadastro_clientes[1]['nome_completo']} irá pagar o valor Total: R$ {vltotal_reservacliente2:.2f} no quarto {carrinho[1]} por {qtdedias_reserva[1]} dias 
    {cadastro_clientes[2]['nome_completo']} irá pagar o valor Total: R$ {vltotal_reservacliente3:.2f} no quarto {carrinho[2]} por {qtdedias_reserva[2]} dias 
""")
print("============================================ Forma de Pagamento ========================================================================================")
if pagamento == 0:
    parcelas = soma / 12
    print(f", o valor total da compra é de R$ {soma:.2f} podera ser parcelado em até 12 vezes sem juros no Cartão de Crédito!!! Sendo 12 parcelas de R$ {parcelas:.2f} ")
elif pagamento == 1:
    print(f""" Como o pagamento será via Pix você terá 10% de desconto em cima do valor total das diárias reservadas e os clientes terão os seguintes descontos:
    {cadastro_clientes[0]['nome_completo']} irá pagar o valor Total: R$ {vltotal_reservacliente1:.2f} sendo o valor final com desconto: R$ {desconto_cliente1:.2f} no quarto {carrinho[0]}     
    {cadastro_clientes[1]['nome_completo']} irá pagar o valor Total: R$ {vltotal_reservacliente2:.2f} sendo o valor final com desconto: R$ {desconto_cliente2:.2f} no quarto {carrinho[1]}  
    {cadastro_clientes[2]['nome_completo']} irá pagar o valor Total: R$ {vltotal_reservacliente3:.2f} sendo o valor final com desconto: R$ {desconto_cliente3:.2f} no quarto {carrinho[2]}
  """)
else:
    print (f""". Infelizmente para o pagamento de Boleto não temos desconto,  o valor total das diárias para cada cliente será de: 
    {cadastro_clientes[0]['nome_completo']} irá pagar o valor Total: R$ {vltotal_reservacliente1:.2f} no quarto {carrinho[0]} 
    {cadastro_clientes[1]['nome_completo']} irá pagar o valor Total: R$ {vltotal_reservacliente2:.2f} no quarto {carrinho[1]}  
    {cadastro_clientes[2]['nome_completo']} irá pagar o valor Total: R$ {vltotal_reservacliente3:.2f} no quarto {carrinho[2]}           
           
           """)


print(f"{cadastro_clientes[0] ["nome_completo"]}, {cadastro_clientes[1] ["nome_completo"]}, {cadastro_clientes[2] ["nome_completo"]} , seus pagamentos foram realizado com sucesso!!!! O  Hotel Time Square - NY SA agradece a preferência e volte sempre!!!!")

```
